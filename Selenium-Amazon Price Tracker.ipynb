```python
driver.get(categories[1])
body_el=driver.find_element_by_css_selector("body")
html_str=body_el.get_attribute("innerHTML")
html_obj= HTML(html=html_str)
```

```python
link="https://www.amazon.in/"
new_links=[x for x in html_obj.links if x.startswith("/")]
print(len(new_links))
```

```python
link="https://www.amazon.in/"
keyword_to_remove=["product-reviews"]
page_links=[link+x for x in html_obj.links if x.startswith("/")]
page_links=[x for x in page_links if 'product-reviews' not in x]

#page_links=[x for x in page_links if '/b/' not in x]

#page_links=[x for x in html_obj.links if i not in x for i in keyword_to_remove]

print(len(page_links))
```

```python
def product_scraping(url):
    driver.get(url)
    body_el=driver.find_element_by_css_selector("body")
    html_str=body_el.get_attribute("innerHTML")
    title_product="#productTitle"
    price='#priceblock_ourprice'
    html_obj= HTML(html=html_str)
    product_title=html_obj.find(title_product,first=True).text
    Pdt_price=html_obj.find(price,first=True).text
    return product_title,Pdt_price
```

```python
pdt_title=[]
pdt_price=[]
pdt_link=[]
for i in page_links:
    
    title,price=(0,0)
    try:
        title,price=product_scraping(i)
    except:
        print(i,"\n")
        pass
    if title!=0 and price!=0:
        pdt_title.append(title)
        pdt_price.append(price)
        pdt_link.append(i)
        print(i)
        print(title,price,"\n")
        
```

**Parse URL using Regex**

```python
#<base-url>/<slug>/dp/product_id>/
import re
reg_pattern=r"https://www.amazon.in//(?P<slug>[\w-]+)/dp/(?P<product_id>[\w-]+)"
url="https://www.amazon.in//Boat-BassHeads-100-Inspired-Earphones/dp/B07GQD4K6L/ref=zg_bs_electronics_47?_encoding=UTF8&psc=1&refRID=EPZABYJS2XP164MVBD6Z"
```

```python
regex=re.compile(reg_pattern)
```

```python
match=regex.match(url)
match["slug"],match["product_id"]
```
